```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
import json

with open('benchmark_results.json') as f:
    data = json.load(f)
```

```python
groups = data['benchmark_results']
```

```python
pairwise_group = [g for g in groups if g['group_name'] == 'pairwise'][0]
```

```python
def plot_group(group, width=0.35):
    records = group['records']
    labels = [r['name'] for r in records]
    warm_time = np.asarray([r['warm_time'] for r in records])
    max_time = np.asarray([r['cold_time'] or r['warm_time'] for r in records])
    overhead = max_time - warm_time
    
    ind = np.arange(len(labels))
    p1 = plt.bar(ind, warm_time,   width, color='g')
    p2 = plt.bar(ind, overhead, width, color='b', bottom=warm_time)
    
    plt.ylabel('Time (s)')
    plt.title(group['group_name'])
    plt.xticks(ind + width /2., labels, rotation=45)
    plt.ylim((0, np.median(max_time) * 2))
    plt.legend( (p1[0], p2[0]), ('Execution time', 'Cold startup overhead'), loc='best')
    plt.savefig(group['group_name'] + '.png')
```

```python
plot_group(pairwise_group)
```
